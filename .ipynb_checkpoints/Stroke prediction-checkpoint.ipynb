#### Discretisation de l'attribut `age`:
D'apres le cdc, la chance d'avoir une AVC double chaque 10 ans, alors, nous allons separer les groupes d'age en intervalle de 10 ans: 0-10, 10-20,... sauf le dernier interval qui est de 100-infinie

```python
def find_age_group(age):
    for i, max_age in enumerate(range(10, 101, 10)):
        if age < max_age: return i
    return 10

dataset_discrete["age_group"] = dataset_raw["age"].transform(find_age_group)
```

#### Discretisation de l'attribut `bmi`:
* certains modele comme naive bayes a besoins de donnees discretes
* On classifie les valeurs de bmi dans des bins: (https://www.cdc.gov/obesity/basics/adult-defining.html)
    - underweight bmi < 18.5,
    - healthy 18.5 <= bmi < 25,
    - overweight 25 <= bmi < 30
    - obese: 30 <= bmi

Tache de discretisation:

```python
def classify_bmi(bmi):
    if np.isnan(bmi): return "unknown"
    if bmi < 18.5: return "Underweight"
    if bmi < 25: return "Healthy"
    if bmi < 30: return "Overweight"
    return "Obese"

# Converting string labels into numbers.
bmi_discrete = le.fit_transform(dataset_raw["bmi"].transform(classify_bmi))
dataset_discrete["bmi_discrete"] = bmi_discrete
```

#### Discretisation de l'attribut `avg_glucose_level`:
* certains modele comme naive bayes a besoins de donnees discretes
* On classifie les valeurs de bmi dans des bins: (https://diabetes.org/diabetes/a1c/diagnosis)
    - does not have diabetes: avg_glucose_level < 200 mg/dl
    - has_diabetes: avg_glucose_level >= 200 mg/dl

Heureusement, notre jeu de donees ne contient pas de valeurs invalides. ce qui facilite la tache de discretisation

```python
dataset_raw["avg_glucose_level"].isna().any()
```

Tache de discretisation:

```python
def classify_is_diabetic(glucose_level):
    return int(glucose_level >= 200)

dataset_discrete["is_diabetic"] = dataset_raw["avg_glucose_level"].transform(classify_is_diabetic)
```

```python
def onehot_encore_and_add_to_dataset(attribute_name):
    global dataset
    rows = dataset_raw[attribute_name].shape[0]
    temp_data = dataset_raw[attribute_name].values.reshape((rows,1))
    n = enc.fit_transform(temp_data);
    rows, cols = n.shape
    temp_data_df = pd.DataFrame(n.toarray(), columns=[f"{attribute_name}_{i}" for i in range(cols)])
    dataset = dataset.join(temp_data_df)
```

#### Encoding de l'attribut `smoking_status` avec One Hot encoding:

```python
onehot_encore_and_add_to_dataset("smoking_status")
```

## Data Preperation (sampling)
On doit separer les donnees du jeu de donnees en 2:
* donnees d'entrainement: afin d'entrainer le modele
* donnees de test: afin de tester la prediction du modele  

Pour le faire on doit utiliser des valeurs diversifier pour les 2, mais surtout de ne pas avoir trop de outliers dans les donnees de test. Pour ce projet, on va prendre 75% des donnees pour l'entrainement et 25% pour le test, mais vu la distribution des resultats (stroke ou non stroke), on fait la distribution de maniere de prendre le meme quotient de resultats (par exemple, sur 10 stroke, on prend 2-3 pour le test).
Afin d'effectuer cette tache, on utilise la fonction `train_test_split` de `sklearn.model_selection` qui separe les donnees de facon "stratified".

```python
target = dataset_raw["stroke"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset, target, test_size=0.25)
```

```python
def print_analysis_report(predictions, y_test):
    report = metrics.classification_report(y_test, predictions)
    print("\n\t\tResult Analysis Report\n", report)
```

## Modeles
### Naive Bayes
#### Algoritme gaussian

```python
def run_gaussian(**kwargs):
    gaussian_model = nb.GaussianNB(**kwargs)
    gaussian_model.fit(X_train, y_train)
    score = gaussian_model.score(X_test, y_test)
    print("Gaussian Algorithm score is", score)
    gaussian_predictions = gaussian_model.predict(X_test)
    print(pd.DataFrame(gaussian_predictions == y_test.values, columns=["results"]).value_counts())
    print_analysis_report(gaussian_predictions, y_test)
```

```python
run_gaussian()
```

### Régression logistique

```python
def run_regression(**kwargs):
    regression_model = lg.LogisticRegression(**kwargs)
    regression_model.fit(X_train, y_train)
    score = regression_model.score(X_test, y_test)
    print("Régression logistique Algorithm score is", score)
    regression_predictions = regression_model.predict(X_test)
    print(pd.DataFrame(regression_predictions == y_test.values, columns=["results"]).value_counts())
    print_analysis_report(regression_predictions, y_test)
```

```python
run_regression()
```

### Perceptron multi-couches

```python
def run_mlp(**kwargs):
    mlp_model = nn.MLPClassifier(**kwargs)
    mlp_model.fit(X_train, y_train)
    score = mlp_model.score(X_test, y_test)
    print("Régression logistique Algorithm score is", score)
    mlp_predictions = mlp_model.predict(X_test)
    ax = pd.DataFrame(mlp_predictions == y_test.values, columns=["results"]).value_counts().plot.bar()
    for container in ax.containers:
        ax.bar_label(container)
    print_analysis_report(mlp_predictions, y_test)
```

```python
run_mlp()
```
