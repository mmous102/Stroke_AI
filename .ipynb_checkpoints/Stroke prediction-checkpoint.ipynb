## Preprocessing
### Gestion de donnees manquante
 On a 2 strategies:
  * enlever les entrees a valeurs manquantes:
      - peut etre une bonne strategie dans le cas de `gender`, car une seule entre est invalide (Other) 
  * remplacer les valeurs manquantes
      - une idee est de remplacer les valeurs de bmi manquante avec la moyenne des bmi presentes (a verifier)
      
### Gestion de type de donnees (discrete/continue)
#### Discretisation de l'attribut `bmi`:
 - certains modele comme naive bayes a besoins de donnees discretes
 - On classifie les valeurs de bmi dans des bins: (https://www.cdc.gov/obesity/basics/adult-defining.html)
     * underweight bmi < 18.5,
     * healthy 18.5 <= bmi < 25,
     * overweight 25 <= bmi < 30
     * obese: 30 <= bmi

Tache de discretisation:

```python
def classify_bmi(bmi):
    if np.isnan(bmi): return "unknown"
    if bmi < 18.5: return "Underweight"
    if bmi < 25: return "Healthy"
    if bmi < 30: return "Overweight"
    return "Obese"

dataset_raw["bmi_discrete"] = dataset_raw["bmi"].transform(classify_bmi)
```

#### Discretisation de l'attribut `avg_glucose_level`:
 - certains modele comme naive bayes a besoins de donnees discretes
 - On classifie les valeurs de bmi dans des bins: (https://diabetes.org/diabetes/a1c/diagnosis)
     * does not have diabetes: avg_glucose_level < 200 mg/dl
     * has_diabetes: avg_glucose_level >= 200 mg/dl

Heureusement, notre jeu de donees ne contient pas de valeurs invalides. ce qui facilite la tache de discretisation

```python
dataset_raw["avg_glucose_level"].isna().any()
```

Tache de discretisation:

```python
def classify_is_diabetic(glucose_level):
    return int(glucose_level >= 200)

dataset_raw["avg_glucose_level_discrete"] = dataset_raw["avg_glucose_level"].transform(classify_is_diabetic)
```

#### Encoding de l'attribut `ever_married`:
Vu que les valeurs de cet attributs sont Yes ou No, on peut juste assigner 0 a No et 1 a Yes.
Afin d'eviter d'ajouter trop d'attributs, on a evite d'utiliser la methode one hot, avec 1 attributs pour chaque valeur discrete 

```python
def encode_ever_married(ever_married):
    return int(ever_married == "Yes")

dataset_raw["ever_married_encoded"] = dataset_raw["ever_married"].transform(encode_ever_married)
```

#### One hot de l'attribut `work_type`:
Les valeurs possibles de work_type sont:

```python
dataset_raw["work_type"].unique().tolist()
```

L'attribut work_type presente 5 differentes valeurs, alors on va utiliser 5 attributs differents pour le representer de facon "continue".
On peut appeler les attributs:
 * `work_type_0`: 1 si work_type est 'Private', sinon 0,
 * `work_type_1`: 1 si work_type est 'Self-employed', sinon 0,
 * `work_type_2`: 1 si work_type est 'Govt_job', sinon 0,
 * `work_type_3`: 1 si work_type est 'children', sinon 0,
 * `work_type_4`: 1 si work_type est 'Never_worked', sinon 0

```python
for i, val in enumerate(dataset_raw["work_type"].unique().tolist()):
    dataset_raw[f"work_type_{i}"] = dataset_raw["work_type"].transform(lambda work_type: int(work_type == val))
```
