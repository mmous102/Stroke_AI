### Gestion de type de donnees (discrete/continue)
#### Discretisation de l'attribut `age`:
 D'apres le cdc, la chance d'avoir une AVC double chaque 10 ans, alors, nous allons separer les groupes d'age en intervalle de 10 ans: 0-10, 10-20,... sauf le dernier interval qui est de 100-infinie

````python
def find_age_group(age):
    for i, max_age in enumerate(range(10, 101, 10)):
        if age < max_age: return i
    return 10

dataset_raw["age_group"] = dataset_raw["age"].transform(find_age_group)
````

#### Discretisation de l'attribut `bmi`:
 - certains modele comme naive bayes a besoins de donnees discretes
 - On classifie les valeurs de bmi dans des bins: (https://www.cdc.gov/obesity/basics/adult-defining.html)
     * underweight bmi < 18.5,
     * healthy 18.5 <= bmi < 25,
     * overweight 25 <= bmi < 30
     * obese: 30 <= bmi

Tache de discretisation:

````python
def classify_bmi(bmi):
    if np.isnan(bmi): return "unknown"
    if bmi < 18.5: return "Underweight"
    if bmi < 25: return "Healthy"
    if bmi < 30: return "Overweight"
    return "Obese"

dataset_raw["bmi_discrete"] = dataset_raw["bmi"].transform(classify_bmi)

# Converting string labels into numbers.
bmi_encoded = le.fit_transform(dataset_raw["bmi_discrete"])
dataset_raw["bmi_encoded"] = bmi_encoded
````

#### Discretisation de l'attribut `avg_glucose_level`:
 - certains modele comme naive bayes a besoins de donnees discretes
 - On classifie les valeurs de bmi dans des bins: (https://diabetes.org/diabetes/a1c/diagnosis)
     * does not have diabetes: avg_glucose_level < 200 mg/dl
     * has_diabetes: avg_glucose_level >= 200 mg/dl

Heureusement, notre jeu de donees ne contient pas de valeurs invalides. ce qui facilite la tache de discretisation

````python
def classify_is_diabetic(glucose_level):
    return int(glucose_level >= 200)

dataset_raw["is_diabetic"] = dataset_raw["avg_glucose_level"].transform(classify_is_diabetic)
````

## Data Preperation (sampling)
On doit separer les donnees du jeu de donnees en 2:
* donnees d'entrainement: afin d'entrainer le modele
* donnees de test: afin de tester la prediction du modele  

Pour le faire on doit utiliser des valeurs diversifier pour les 2, mais surtout de ne pas avoir trop de outliers dans les donnees de test. Pour ce projet, on va prendre 75% des donnees pour l'entrainement et 25% pour le test, mais vu la distribution des resultats (stroke ou non stroke), on fait la distribution de maniere de prendre le meme quotient de resultats (par exemple, sur 10 stroke, on prend 2-3 pour le test).
Afin d'effectuer cette tache, on utilise la fonction `train_test_split` de `sklearn.model_selection` qui separe les donnees de facon "stratified".

On execute le code suivant afin d'avoir 2 echantillons. Pour garder nos donnees consistantes pendant les tests, on exporte les dataframes en 2 fichiers csv.
